```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
```

# Cargamos los datos:

```python
df= pd.read_csv("NY-House-Dataset.csv")
df.head()
```

# Limpiamos y organizamos el DataFrame:

```python
df1= df[['TYPE', 'PRICE', 'BEDS','BATH','PROPERTYSQFT', 'STATE', 'LOCALITY']]
df1
```

```python
df1.info()
```

```python
df1= df1.astype({'BATH': int, 'PROPERTYSQFT':int})
df1
```

```python
df1.rename(columns={'TYPE':'tipo', 'PRICE':'precio', 'BEDS':'habitaciones','BATH':'baños','PROPERTYSQFT':'tamaño(ft)', 'STATE':'estado', 'LOCALITY': 'localidad'},inplace=True)
df1
```

```python
df1['estado']= df1['estado'].str.split(',').str.get(0)
df1
```

```python
df1['tipo']= df1['tipo'].str.split(' ').str.get(0)
df1
```

```python
df1['tamaño(m2)']= df1['tamaño(ft)'] / 3
df1
```

```python
df1.info()
```

```python
df1['tamaño(m2)']= df1['tamaño(m2)'].astype(int) 
df1.info()
```

```python
df1.pop('tamaño(ft)')
df1
```

```python
df1['prec/m2']= df1['precio'] / df1['tamaño(m2)']
df1
```

```python
df1['prec/m2']= df1['prec/m2'].astype(int)
```

```python
df1.info()
```

# Graficar todos los tipos de casas menores de 1000 m2 y menos de 1M $:

```python
dfmil= df1[(df1['precio'] <= 1000000) & (df1['tamaño(m2)'] <= 1000)]
dfmil
```

```python
sns.set_theme(style='white', context='talk')
graf=sns.relplot(data= dfmil, height= 7, aspect= 1.3,
                          x='tamaño(m2)',
                          y= 'precio',
                          hue='tipo',
                          sizes=(100,500),
                          style='tipo')
graf.set_xlabels('Tamaño(m2)')
graf.set_ylabels('Precio(M$)')
graf.set(title= 'Tipos de Casas de menos de 1M$ y menos de 1000 m2')
plt.grid()
plt.xticks(np.arange(0,1001, 100))
plt.xticks(np.arange(0, 1001, 50), rotation= 270)
plt.yticks(np.arange(0,1000001, 50000))
plt.minorticks_on()

plt.show()
```

```python
## Observamos los datos más importantes de cada tipo:
#    -Condo: La mayoría se encuentran en un intervalo de menos de 400 m2 y de entre 400.000 y 1M de $.
#    -House: Podemos encontrarlas de todos los precios y tamaños, siendo mayoría las de más de 500.000 $.
#    -Co-op: La mayoría son de menos de 400 m2 y menos de 500.000 $, pero las podemos encontrar de 730 m2 a cualquier precio.
#    -Townhouse: Se pueden encontrar entre 400 y 800 m2 y entre 300.000 y 1M $.
#    -Multi-family: La mayoría son de más de 650.000$ y mayores de 450 m2.
#    -Contingent: Podemos encontrarlos pequeños y baratos, grandes y caros y pequeños y muy caros.
#    -Land: La mayoría son de unos 730 m2 de casi todos los precios. Hay 2 por menos de 100.000$ y una de unos 300 m2 y 1 M$.
#    -Foreclosure: Hay poca oferta y van desde los 250.000 hasta los 900.000 $, y los tamaños van entre los 250 y los 900 m2.
#    -For: Las hay desde muy pequeñas y casi gratis, muy pequeñas y muy caras y de 730 m2 de varios precios.
#    -Coming: Sólo tenemos una oferta de 730 m2 y 650.000 $.
#    -Pending: Los hay de todos los tamaños y precios y en proporción entre ambos.
#    -Condop: Sólo tenemos uno y está a 600.000 $ de 730 m2.
```

# Graficamos la cantidad de ofertas que hay de cada tipo:

```python
dfm= dfmil['tipo'].value_counts()
dfm
```

```python
plt.pie(dfm, labels=['Co-op', 'House', 'Condo',  'Multi-family', 'Pending', 'Contingent', 'Townhouse', 'Land', 'Foreclosure', 'For', 'Coming', 'Condop'], radius=1.2, labeldistance= 0.6)
plt.title('Cantidad de edificaciones de cada tipo')
```

# Graficar todos los tipos de casas con menos de 6000 $ el m2 en cada estado:

```python
dfm2= df1[df1['prec/m2'] < 6000]
dfm2
```

```python
sns.set_theme(style='white', context='talk')
graf=sns.relplot(data= dfm2, height= 7, aspect= 1.8,
                          x='estado',
                          y= 'prec/m2',
                          hue='tipo',
                          sizes=(100,500),
                          style='tipo')
graf.set_xlabels('Estado')
graf.set_ylabels('Precio/m2 ($)')
graf.set(title= 'Precio del m2 de cada tipo en cada estado')
plt.grid()
#plt.xticks(np.arange(0,3501, 100))
plt.xticks(rotation= 270)
plt.yticks(np.arange(0,6001, 200))
plt.minorticks_on()

plt.show()
```

# Sacar las casas de NY con entre 2 y 4 habitaciones, y menos de 1M $

```python
casasNY= df1[(df1['tipo'] == 'House') & (df1['localidad'] == 'New York') & (df1['habitaciones'] >= 2) & (df1['habitaciones'] <= 4) & (df1['precio'] <= 1e6)]
casasNY
```

```python
sns.set_theme(style= 'darkgrid', context= 'talk')
sns.relplot(data=casasNY, height= 7, aspect=1.1, x='precio', y='tamaño(m2)', hue= 'habitaciones')
plt.title('Casas en NY de cada precio y tamaño por habitaciones')
plt.yticks(range(0, 1400, 50))
plt.ylabel('Tamaño(m2)')
plt.xlabel('Precio($)')
```
